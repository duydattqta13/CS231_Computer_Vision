```python
import matplotlib.pyplot as plt

def _showHist(pathfilename):
  img = cv.imread(pathfilename,0)
  hist = cv.calcHist([img],[0],None,
  [256],[0,256])
  size = img.shape[0]*img.shape[1]
  hist = hist / size

  plt.plot(hist)
  plt.xlim([0,256])
  plt.legend(('histogram'),loc = 'upper left')
  plt.show()

_showHist(path_filename)
```

# 2.Tính khoảng cách Euclid (Scipy) và sự tương đồng dùng cosine


```python
def TinhHist(pathfilename):
  img = cv.imread(pathfilename,0)
  hist = cv.calcHist([img],[0],None,
  [256],[0,256])
  size = img.shape[0]*img.shape[1]
  hist = hist / size
  return hist
```

```python
from scipy.spatial import distance

path_filename_0814 = os.path.join(dir_images, 'image_0814.jpg')
path_filename_0816 = os.path.join(dir_images, 'image_0816.jpg')
path_filename_1248 = os.path.join(dir_images, 'image_1248.jpg')

image_0814 = TinhHist(path_filename_0814)
image_0816 = TinhHist(path_filename_0816)
image_1248 = TinhHist(path_filename_1248)
```

-Khoảng cách Euclid 

```python
print('Euclid distance image_0814 and image_0816: ' ,distance.euclidean(image_0814, image_0816))
print('Euclid distance image_0814 and image_1248: ' ,distance.euclidean(image_0814, image_1248))
```

-Cosine Similarity

```python
print('Cosine Similarity image_0814 and image_0816: ',1 - distance.cosine(image_0814, image_0816))
print('Cosine Similarity image_0814 and image_1248: ',1 - distance.cosine(image_0814, image_1248))
```
